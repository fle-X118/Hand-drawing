```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np


df = sns.load_dataset("tips")
df["sex"] = df["sex"].map({"Male" : 0, "Female" : 1})
df["smoker"] = df["smoker"].map({"No" : 0, "Yes" : 1})
df['time'] = df['time'].map({'Dinner': 1, 'Lunch': 0})
df = pd.get_dummies(df, columns=["day"], drop_first=True, dtype=int)
df
```

```python
sns.lmplot(data=df, x="total_bill", y="tip", hue="sex", col="smoker")
plt.show()
```

```python
x = df.drop(["tip"], axis=1)
y = df["tip"]
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.25,
    random_state=69
)
model = RandomForestRegressor(n_estimators=1488, random_state=67)
model2 = LinearRegression()
model2.fit(x_train, y_train)
fuf2 = model2.score(x_test, y_test)
print(f"LinearRegression : {fuf2}")
model.fit(x_train, y_train)
fuf = model.score(x_test, y_test)
print(f"RFR : {fuf}")
```
